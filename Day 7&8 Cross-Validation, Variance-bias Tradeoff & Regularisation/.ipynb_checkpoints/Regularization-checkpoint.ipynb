```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet

%matplotlib inline
```

```python
df = pd.read_csv(r'test1.csv', names = ['X1', 'X2', 'Y'])
```

```python
df
```

```python
model = LinearRegression()
model.fit(df[['X1']], df['Y'])
```

```python
model.coef_
```

```python
model.intercept_
```

```python
df.plot(kind = 'scatter', x= 'X1', y= 'Y')
```

```python
pred_df = model.predict(df[['X1']])
pred_df
```

```python
plt.scatter(df.X1, df.Y, color = 'red')
plt.plot(df.X1, pred_df, 'b')
```

```python
model = LinearRegression()
model.fit(df[['X1']], df['Y'])

print('Slope: ', model.coef_)
print('Intercept: ', model.intercept_)

df.plot(kind = 'scatter', x= 'X1', y= 'Y')

pred_df = model.predict(df[['X1']])
pred_df

plt.scatter(df.X1, df.Y, color = 'red')
plt.plot(df.X1, pred_df, 'b')
```

```python
Lasso?
```

```python
model = Ridge(alpha=0.0005)
model.fit(df[['X1']], df['Y'])

print('Slope: ', model.coef_)
print('Intercept: ', model.intercept_)

df.plot(kind = 'scatter', x= 'X1', y= 'Y')

pred_df = model.predict(df[['X1']])
pred_df

plt.scatter(df.X1, df.Y, color = 'red')
plt.plot(df.X1, pred_df, 'b')
```

```python
model = Lasso(alpha=0.04)
model.fit(df[['X1']], df['Y'])

print('Slope: ', model.coef_)
print('Intercept: ', model.intercept_)

df.plot(kind = 'scatter', x= 'X1', y= 'Y')

pred_df = model.predict(df[['X1']])
pred_df

plt.scatter(df.X1, df.Y, color = 'red')
plt.plot(df.X1, pred_df, 'b')
```

```python
for i in np.arange(0, 0.05, 0.001):
    print(i)
    model = Lasso(alpha= i)
    model.fit(df[['X1']], df['Y'])

    print('Slope: ', model.coef_)
    print('Intercept: ', model.intercept_)

    df.plot(kind = 'scatter', x= 'X1', y= 'Y')

    pred_df = model.predict(df[['X1']])
    pred_df

    plt.scatter(df.X1, df.Y, color = 'red')
    plt.plot(df.X1, pred_df, 'b')
```

```python
model = Ridge(alpha=0.25)
model.fit(df[['X1']], df['Y'])

print('Slope: ', model.coef_)
print('Intercept: ', model.intercept_)

df.plot(kind = 'scatter', x= 'X1', y= 'Y')

pred_df = model.predict(df[['X1']])
pred_df

plt.scatter(df.X1, df.Y, color = 'red')
plt.plot(df.X1, pred_df, 'b')
```

```python
ElasticNet?
```

```python
model = ElasticNet(l1_ratio=0.2, alpha=0.003)
model.fit(df[['X1']], df['Y'])

print('Slope: ', model.coef_)
print('Intercept: ', model.intercept_)

df.plot(kind = 'scatter', x= 'X1', y= 'Y')

pred_df = model.predict(df[['X1']])
pred_df

plt.scatter(df.X1, df.Y, color = 'red')
plt.plot(df.X1, pred_df, 'b')
```
